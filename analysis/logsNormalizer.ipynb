```python
import pandas as pd
import numpy as np
from IPython.display import clear_output

df = pd.read_csv('data/clean.csv')
display(df.shape)
```

```python
servers = sorted(['java', 'nodejs'])
databases = sorted(['nosql', 'sql'])
instances = sorted(['b1', 'b2', 'b4', 'b8'])

combinations = [(servers[0], databases[0]), (servers[1], databases[0]), (servers[0], databases[1]), (servers[1], databases[1])]
sizes = sorted(df['requestSize'].unique())
frequencies = sorted(df['frequency'].unique())

type_combinations = np.transpose([np.tile(sizes, len(frequencies)), np.repeat(frequencies, len(sizes))])
```

```python
total_counts = 0
entries = []

for instance in instances:
    for type_combination in type_combinations:
        size, frequency = type_combination
        
        clear_output()
        display('{} {} {}'.format(instance, size, frequency))

        counts = []        

        for combination in combinations:
            server, database = combination    
            
            partial = df.loc[(df['instanceType'] == instance) & (df['serverType'] == server) & (df['databaseType'] == database) & (df['requestSize'] == size) & (df['frequency'] == frequency)]
            partial_n = partial.shape[0]
            
            if partial_n > 0:
                counts.append(partial_n)
                total_counts = total_counts + partial_n
        
        if len(counts) < 1:
            continue
        
        min_count = min(counts)
        
        for combination in combinations:
            server, database = combination                
            
            partial = df.loc[(df['instanceType'] == instance) & (df['serverType'] == server) & (df['databaseType'] == database) & (df['requestSize'] == size) & (df['frequency'] == frequency)]
            partial_n = partial.shape[0]
            
            if partial_n > 0 and min_count > 0:
                entries = entries + partial.tail(min_count).to_dict('records')
```

```python
print(total_counts, len(entries))
new_df = pd.DataFrame(data=entries)
new_df.to_csv('data/normalized.csv', index=False)
new_df
```
